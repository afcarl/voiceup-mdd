```python
r = requests.get('https://voicesurvey.mit.edu/sites/default/files/documents/campaign.json')
questions = r.json()
```

```python
question_type_mapper = dict(single_choice="irk-text-choice-question-step", audio="irk-audio-task")
mappings = dict(question_prompt_id="id", question_text="text")

tasks = {}
idmap = {}
for question in questions['questions']:
    id = question['question_prompt_id']
    '''
    if id in tasks:
        print('Repeated id: {}'.format(id))
        continue
    '''
    question_type = question['question_type']
    if question_type not in question_type_mapper:
        print('unknown question type: {0} - {1}'.format(question['question_id'], question))
        continue
    idmap[question['question_id']] = id
    task_type = question_type_mapper[question_type]
    if 'single_choice' in question_type:
        text = 'Select only one'
        if 'png' in question['choices'][0]:
            print(id)
            task_type = "irk-image-choice-quetion-step"
            choices = [{"normal-state-image": val.split('(')[-1].split(')')[0], "value": idx + 1} for idx, val in enumerate(question['choices'])]
        else:
            choices = [{"text": val, "value": idx + 1} for idx, val in enumerate(question['choices'])]
        steps = [dict(type=task_type, title=question['question_text'], text=text, choices=choices)]
    elif 'audio' in question_type:
        pre_task = {"type": "instruction",
                    "title": "Record voice",
                    "text": "On the following page you will be asked to read a sentence and record your voice. Please click the record button to start recording. You can continue after the recording has finished.",
                    "image": "irk-voice-recording",
                    "footer-attach": "true"}
        passage = question['question_text']
        if passage.startswith('Read: '):
            passage = passage.split('Read: ')[1]
        audio_task = dict(type=task_type, text=passage, 
                          duration=int(np.ceil(len(passage.split()) * 0.5)))
        steps = [pre_task, audio_task]
    tasks[id] = dict(steps=steps)
    if 'PHQ' in id:
        tasks[id]['timelimit'] = '7-0' 
tasks = dict(tasks=tasks)
```

```python
surveys = dict(surveys={})
all_tasks = []
for survey in questions['surveys']:
    id = survey['id']
    day = int(survey['date'].split('dayOfMonth')[1])
    '''if day > 7:
        continue
    '''
    date = '0 3 {} * *'.format(day)
    task_ids = []
    for val in survey['questions']:
        if idmap[val] == 'SAGE_3': # do at least 1 picture question
            task_ids.append('SAGE_3p')
        else:
            task_ids.append(idmap[val])
    surveys['surveys'][id] = dict(date=date, tasks=task_ids, skippable=task_ids)
    for task_id in task_ids:
        if task_id not in all_tasks:
            all_tasks.append(task_id)

keys = tasks['tasks'].keys()

for key in keys:
    if key not in all_tasks:
        del tasks['tasks'][key]
```

```python
full_dict = {}
full_dict.update(metadata)
full_dict.update(profile)
full_dict.update(eligibility)
full_dict.update(consent)
full_dict.update(tasks)
full_dict.update(surveys)
print(json.dumps(full_dict, sort_keys=True, indent=4))
with open('../specs/spec_{}.json'.format(datestr), 'wt') as fp:
    json.dump(full_dict, fp, sort_keys=True, indent=2)
```
